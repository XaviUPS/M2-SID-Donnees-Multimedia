# 1. Création des CSV (video_id, label)

```python
corpus_dir = "/content/drive/MyDrive/Projet non-alternant/Corpus/json"
train_json = os.path.join(corpus_dir, "train_videodatainfo_audio.json")
val_json = os.path.join(corpus_dir, "val_videodatainfo_audio.json")
test_json = os.path.join(corpus_dir, "test_videodatainfo_audio.json")
```

```python
def extract_video_labels(json_path):
    with open(json_path, 'r') as f:
        data = json.load(f)
    records = []
    for video in data["videos"]:
        records.append({
            "video_id": video["video_id"],
            "label": video["category"]
        })
    return records
```

```python
train_records = extract_video_labels(train_json)
val_records = extract_video_labels(val_json)
test_records = extract_video_labels(test_json)

df_train = pd.DataFrame(train_records)
df_val = pd.DataFrame(val_records)
df_test = pd.DataFrame(test_records)

output_dir = "/content/drive/MyDrive/Projet non-alternant/Corpus/csv/"
output_csv_train = os.path.join(output_dir, "train_audio_labels.csv")
output_csv_val = os.path.join(output_dir, "val_audio_labels.csv")
output_csv_test = os.path.join(output_dir, "test_audio_labels.csv")

df_train.to_csv(output_csv_train, index=False)
df_val.to_csv(output_csv_val, index=False)
df_test.to_csv(output_csv_test, index=False)

print(f"\n✅ CSV sauvegardé ici : {output_dir}")
```
