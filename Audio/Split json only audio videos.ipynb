```python
import os
import json
from natsort import natsorted
from google.colab import drive
```

```python
drive.mount("/content/drive", force_remount=True)
```

```python
audio_dir_train_val = "/content/drive/MyDrive/Projet non-alternant/Corpus/train_val_videos/TrainValAudio/"
audio_dir_test = "/content/drive/MyDrive/Projet non-alternant/Corpus/test_videos/TestAudio/"


audio_files_train_val = [f for f in os.listdir(audio_dir_train_val) if f.endswith(".wav")]
audio_files_test = [f for f in os.listdir(audio_dir_test) if f.endswith(".wav")]

audio_files_train_val = natsorted(audio_files_train_val)
audio_files_test = natsorted(audio_files_test)

audio_video_ids_train_val = {os.path.splitext(os.path.basename(f))[0] for f in audio_files_train_val}
audio_video_ids_test = {os.path.splitext(os.path.basename(f))[0] for f in audio_files_test}

print(f"Found {len(audio_video_ids_train_val)} train audio files.")
print(f"Found {len(audio_video_ids_test)} test audio files.")
```

```python
def filter_json_by_audio(input_path, output_path, audio_video_ids):
    with open(input_path, 'r') as f:
        data = json.load(f)

    filtered_videos = [v for v in data['videos'] if v['video_id'] in audio_video_ids]
    filtered_video_ids = {v['video_id'] for v in filtered_videos}
    filtered_sentences = [s for s in data['sentences'] if s['video_id'] in filtered_video_ids]

    filtered_data = {
        "info": data["info"],
        "videos": filtered_videos,
        "sentences": filtered_sentences
    }

    with open(output_path, 'w') as f:
        json.dump(filtered_data, f, indent=2)

    print(f"{input_path} → {output_path}")
    print(f"  Kept {len(filtered_videos)} videos and {len(filtered_sentences)} captions.")
```

```python
chemin_jsons : str = "/content/drive/MyDrive/Projet non-alternant/Corpus/json/"

filter_json_by_audio(chemin_jsons + 'train_videodatainfo.json', chemin_jsons + 'train_videodatainfo_audio.json', audio_video_ids_train_val)
filter_json_by_audio(chemin_jsons + 'val_videodatainfo.json', chemin_jsons + 'val_videodatainfo_audio.json', audio_video_ids_train_val)
filter_json_by_audio(chemin_jsons + 'test_videodatainfo.json', chemin_jsons + 'test_videodatainfo_audio.json', audio_video_ids_test)
```
