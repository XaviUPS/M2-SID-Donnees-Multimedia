Conversion des videos.mp4 à des audios.wav
==============
---

```python
import os
from natsort import natsorted
from google.colab import drive
from moviepy.editor import VideoFileClip
```

```python
drive.mount("/content/drive", force_remount=True)
```

```python
input_folder = "/content/drive/MyDrive/Projet non-alternant/Corpus/train_val_videos/TrainValVideo"
output_folder = "/content/drive/MyDrive/Projet non-alternant/Corpus/train_val_videos/TrainValAudio"

input_folder = "/content/drive/MyDrive/Projet non-alternant/Corpus/test_videos/TestVideo"
output_folder = "/content/drive/MyDrive/Projet non-alternant/Corpus/test_videos/TestAudio"
```

```python
mp4_files = [f for f in os.listdir(input_folder) if f.endswith(".mp4")]
mp4_files = natsorted(mp4_files)

print(f"Found {len(mp4_files)} MP4 files.")
```

```python
for filename in mp4_files:
    input_path = os.path.join(input_folder, filename)
    output_filename = os.path.splitext(filename)[0] + ".wav"
    output_path = os.path.join(output_folder, output_filename)

    print(f"Converting {filename} → {output_filename} ...")

    try:
        video = VideoFileClip(input_path)
        if video.audio is None:
            print(f"    ⚠️  No audio in {filename}. Skipping.")
            video.close()
            continue

        video.audio.write_audiofile(
            output_path,
            fps=22050,
            codec='pcm_s16le',
            nbytes=2,
            verbose=False,
            logger=None
        )
        video.close()

    except Exception as e:
        print(f"    ❌ Error with {filename}: {e}")

print("✅ Conversion completed!")
```

```python
wav_files = [f for f in os.listdir(output_folder) if f.endswith(".wav")]
wav_files = natsorted(wav_files)

print(f"Found {len(wav_files)} MP4 files.")
```
