```python
from google.colab import drive


drive.mount("/content/drive", force_remount=True)
input_folder = "/content/drive/MyDrive/Projet non-alternant/Corpus/"
```

```python
import json


with open(input_folder + 'train_val_videodatainfo.json', 'r') as f:
    data = json.load(f)

info = data['info']
videos = data['videos']
sentences = data['sentences']
```

```python
def get_video_num(vid_str):
    return int(vid_str.replace('video', ''))


train_max_id = 6512
val_min_id = 6513
val_max_id = 7009

train_videos = [v for v in videos if get_video_num(v['video_id']) <= train_max_id]
val_videos = [v for v in videos if val_min_id <= get_video_num(v['video_id']) <= val_max_id]

train_video_ids = {v['video_id'] for v in train_videos}
val_video_ids = {v['video_id'] for v in val_videos}

train_sentences = [s for s in sentences if s['video_id'] in train_video_ids]
val_sentences = [s for s in sentences if s['video_id'] in val_video_ids]

train_data = {
    "info": info,
    "videos": train_videos,
    "sentences": train_sentences
}

val_data = {
    "info": info,
    "videos": val_videos,
    "sentences": val_sentences
}
```

```python
with open(input_folder + 'train_videodatainfo.json', 'w') as f:
    json.dump(train_data, f, indent=2)

with open(input_folder + 'val_videodatainfo.json', 'w') as f:
    json.dump(val_data, f, indent=2)

print("Split completed!")
print(f"Train: {len(train_videos)} videos, {len(train_sentences)} captions")
print(f"Val: {len(val_videos)} videos, {len(val_sentences)} captions")
```
