```python
from google.colab import drive
import json

# --- Monter Google Drive ---
drive.mount("/content/drive", force_remount=True)

# --- D√©finir le chemin du dossier ---
chemin_videos = '/content/drive/MyDrive/Projet non-alternant/Corpus/json/'

# --- Ouvrir le fichier JSON ---
json_path = chemin_videos + 'test_videodatainfo.json'

with open(json_path, 'r') as f:
    metadata = json.load(f)

metadata = {item['video_id']: item for item in metadata['videos']}
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
from google.colab import drive
# a) Indexation des images
# Assuming the 'quantification' and 'histogramme' functions are defined in previous cells
drive.mount("/content/drive", force_remount=True)
chemin_videos = 'drive/MyDrive/Projet non-alternant/Corpus/train_val_videos/TrainValVideo/'
```

```python
#transformation des vid√©os en images
import cv2
import os
import pandas as pd
import numpy as np

rows = []

# Liste des vid√©os
fichiers_videos = sorted([
    f for f in os.listdir(chemin_videos)
    if f.lower().endswith('.mp4')
])




# Limite aux vid√©os 2000 √† 4000
fichiers_videos = fichiers_videos[:2000]

for video_file in fichiers_videos:
    video_path = os.path.join(chemin_videos, video_file)
    cap = cv2.VideoCapture(video_path)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    if total_frames < 16:
        cap.release()
        continue  # on saute les vid√©os trop courtes

    # 16 indices uniform√©ment espac√©s sur toute la vid√©o
    interval = total_frames / 64
    uniform_indices = [int(i * interval) for i in range(16)]

    frames = []
    for idx in uniform_indices:
        cap.set(cv2.CAP_PROP_POS_FRAMES, idx)
        ret, frame = cap.read()
        if ret:
            frame = cv2.resize(frame, (224, 224))  # redimension facultatif
            frames.append(frame)

    cap.release()


    video_id = video_file.replace(".mp4", "")
    category = metadata.get(video_id, {}).get("category", None)

    for i, img in enumerate(frames):
        rows.append({
            "video_name": video_file,
            "category": category,
            "image_index": uniform_indices[i],
            "image_pixels": img
        })

    del frames

# Cr√©ation du DataFrame final
df_images = pd.DataFrame(rows)

print("\n‚úÖ DataFrame cr√©√© avec 16 frames uniformes par vid√©o :")
print(df_images.head())
```

```python
# ============================================================
# 1Ô∏è‚É£ Imports
# ============================================================
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import models, transforms
from torch.utils.data import DataLoader, Dataset
import torch.nn.functional as F
import cv2
import numpy as np
from tqdm import tqdm
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
# ============================================================
# 2Ô∏è‚É£ Dataset vid√©o (adapt√© √† vos frames)
# ============================================================
videos = []
labels = []

for video_name, group in df_images.groupby("video_name"):
    group = group.sort_values("image_index")
    frames = [img for img in group["image_pixels"].tolist()]
    if len(frames) == 16:  # 16 frames par vid√©o
        videos.append(np.stack(frames))  # (16, H, W, 3)
        labels.append(group["category"].iloc[0])
    else:
        continue

le = LabelEncoder()
labels = le.fit_transform(labels)
```

```python
import pickle
import cv2
import numpy as np

# Fichiers √† traiter
x_files = [
    "/content/drive/MyDrive/Projet non-alternant/Vid√©o/ensemble_non_hasard/X_train1.pkl",
    "/content/drive/MyDrive/Projet non-alternant/Vid√©o/ensemble_non_hasard/X_train2.pkl",
    "/content/drive/MyDrive/Projet non-alternant/Vid√©o/ensemble_non_hasard/X_train3.pkl",
    "/content/drive/MyDrive/Projet non-alternant/Vid√©o/ensemble_non_hasard/X_train4.pkl",
]

x_val_files = [
    "/content/drive/MyDrive/Projet non-alternant/Vid√©o/ensemble_non_hasard/X_val1.pkl",
    "/content/drive/MyDrive/Projet non-alternant/Vid√©o/ensemble_non_hasard/X_val2.pkl",
    "/content/drive/MyDrive/Projet non-alternant/Vid√©o/ensemble_non_hasard/X_val3.pkl",
    "/content/drive/MyDrive/Projet non-alternant/Vid√©o/ensemble_non_hasard/X_val4.pkl",
]

# Nouvelle r√©solution
new_size = (112, 112)

def process_pickle(file_path):
    # Charger pickle
    with open(file_path, "rb") as f:
        videos = pickle.load(f)

    processed_videos = []
    for video in videos:
        new_video = []
        for frame in video:
            # Convertir en uint8 si n√©cessaire
            frame = frame.astype(np.uint8)

            # Redimensionner
            frame_resized = cv2.resize(frame, new_size)

            new_video.append(frame_resized)
        processed_videos.append(new_video)

    # Sauvegarder avec le m√™me nom ou ajouter "_small"
    new_file = file_path.replace(".pkl", "_small3.pkl")
    with open(new_file, "wb") as f:
        pickle.dump(processed_videos, f, protocol=4)
    print(f"‚úÖ Fichier trait√© et sauvegard√© : {new_file}")

# Traiter tous les fichiers
for f in x_files + x_val_files:
    process_pickle(f)
```
