```python

import pandas as pd
import numpy as np

from google.colab import drive

# ============================================================
# 2ï¸âƒ£ MONTAGE DU DRIVE ET CHARGEMENT DU DATAFRAME
# ============================================================
drive.mount("/content/drive", force_remount=True)

save_path = '/content/drive/MyDrive/Projet non-alternant/VidÃ©o/df_images.pkl'
df_images = pd.read_pickle(save_path)
print("âœ… DataFrame chargÃ© :", df_images.shape)
videos = []
labels = []

for video_name, group in df_images.groupby("video_name"):
    group = group.sort_values("image_index")
    frames = [img for img in group["image_pixels"].tolist()]
    if len(frames) == 16:  # 16 frames par vidÃ©o
        videos.append(np.stack(frames))  # (16, H, W, 3)
        labels.append(group["category"].iloc[0])
    else:
        continue

import pickle

save_dataset_path = '/content/drive/MyDrive/Projet non-alternant/VidÃ©o/resnet_images.pkl'

with open(save_dataset_path, 'wb') as f:
    pickle.dump((videos, labels), f)

print("âœ… Dataset sauvegardÃ© :", save_dataset_path)
```

```python
import pickle
import pandas as pd
import numpy as np

from google.colab import drive

# ============================================================
# 2ï¸âƒ£ MONTAGE DU DRIVE ET CHARGEMENT DU DATAFRAME
# ============================================================
drive.mount("/content/drive", force_remount=True)

# ðŸ“ Dossier oÃ¹ sont stockÃ©s tes fichiers
base_path = '/content/drive/MyDrive/Projet non-alternant/VidÃ©o/'

# ============================================================
# ðŸ”¹ Fonctions utilitaires
# ============================================================
import pickle

base_path = '/content/drive/MyDrive/Projet non-alternant/VidÃ©o/'
batch_size = 8000  # ajuste selon ta RAM
merged = []
part_num = 1

for i in range(1, 4):
    with open(f"{base_path}X_train{i}.pkl", 'rb') as f:
        data = pickle.load(f)
    for d in data:
        merged.append(d)
        if len(merged) >= batch_size:
            out_path = f"{base_path}X_train_full_part{part_num}.pkl"
            with open(out_path, 'wb') as fout:
                pickle.dump(merged, fout)
            print(f"ðŸ’¾ SauvegardÃ© {out_path} ({len(merged)} Ã©lÃ©ments)")
            merged = []
            part_num += 1

# Sauvegarde du dernier lot
if merged:
    out_path = f"{base_path}X_train_full_part{part_num}.pkl"
    with open(out_path, 'wb') as fout:
        pickle.dump(merged, fout)
    print(f"ðŸ’¾ SauvegardÃ© {out_path} ({len(merged)} Ã©lÃ©ments)")

```
